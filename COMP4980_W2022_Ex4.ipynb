```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_diabetes

diab = load_diabetes()
```

```python
diab.feature_names
```

```python
X = diab.data
y = diab.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X[:,8].reshape(-1,1), y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
```

The r2 score of my linear regression model using features 2 and 8 is 0.418112859

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
# only using feature 8
plt.figure(figsize=(10,5))
plt.scatter(X_test,y_test,color='red',label='data')
plt.plot(X_test,y_pred,color='black',label = 'pred')
plt.legend()
plt.show()
```
