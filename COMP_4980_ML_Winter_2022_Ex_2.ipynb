```python
r = { 'a' : np.arange(50),
      'b' : np.random.randint(0,10,50),
      'c' : 10*np.random.randn(50)}
r['d'] = r['a'] + 10 * np.random.rand(50)

plt.scatter('a', 'd', s = 'b', c = 'c', data = r)
plt.colorbar()
plt.show()
```

```python
def minMaxScaler(s):
  mini = min(s)
  rng = max(s) - mini

  return [ (n-mini)/rng for n in s  ]
```

```python
r2 = r
for k in r2.keys():
  r2[k] = minMaxScaler(r2[k])

plt.scatter('a', 'd', s = 'b', c = 'c', data = r2)
plt.colorbar()
plt.show()
```

```python
dice = { 'rolls' : np.random.randint(1,7,100) + np.random.randint(1,7,100)}

dice['sums'], dice['count'] = np.unique( dice['rolls'], return_counts= True )

plt.subplot(212)
plt.bar('sums', 'count', data=dice)

plt.subplot(211)
plt.hist(dice['rolls'], bins=5, color=['green'])
```
