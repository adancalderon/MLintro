```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_california_housing
ds = fetch_california_housing()
X = ds.data
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
#scaler = StandardScaler()
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
cls = KMeans(n_clusters=5)
cls.fit(X)
y_pred = cls.predict(X)
```

```python
n = X.shape[1]
plt.figure(figsize=[8, 3*n])

for i in range(n//2):
  plt.subplot(n//2, 2, i+1)
  plt.scatter(X[:, i*2+1], X[:, i*2], c=y_pred)
  plt.xlabel(ds.feature_names[i*2+1])
  plt.ylabel(ds.feature_names[i*2])

plt.colorbar()
  
```

```python
results = []
for k in range(2,11):
  cls = KMeans(n_clusters=k)
  cls.fit(X)
  #y_pred = cls.predict(X)
  results.append(-1*cls.score(X))

plt.plot(list(range(2,11)), results)
```
